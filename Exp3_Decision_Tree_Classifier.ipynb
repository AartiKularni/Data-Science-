```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# 1. Sample Dataset
data = {
    'Credit_Score': ['Poor', 'Fair', 'Good', 'Good', 'Fair', 'Poor', 'Good', 'Fair', 'Poor', 'Good'],
    'Income': [25000, 40000, 60000, 80000, 30000, 20000, 75000, 35000, 15000, 90000],
    'Loan_Amount': [10000, 15000, 20000, 25000, 12000, 8000, 22000, 13000, 7000, 24000],
    'Approved': ['No', 'Yes', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

# 2. Encode categorical variables
df['Credit_Score'] = df['Credit_Score'].map({'Poor': 0, 'Fair': 1, 'Good': 2})
df['Approved'] = df['Approved'].map({'No': 0, 'Yes': 1})

# 3. Features and Labels
X = df[['Credit_Score', 'Income', 'Loan_Amount']]
y = df['Approved']

# 4. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Train Decision Tree
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)

# 6. Predict and Evaluate
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# 7. Plot the Tree
plt.figure(figsize=(10,9))
plot_tree(clf, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.title("Decision Tree for Loan Approval")
plt.show()
```
